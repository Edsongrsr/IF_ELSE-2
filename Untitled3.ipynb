```python
entrega = input("Digite o tipo de entrega :")
if entrega == "padrão":
  print("Entrega padrão : r$10,00")
elif entrega == "expresso":
  print("entrega expresso : r$ 20,00")
else :
  print("opção de entrega inválida")
```

```python
#atividade 01
numero = float(input("Digite um número: "))
if numero > 0:
  print("número positivo")
elif numero == 0:
  print("número nulo")
else:
  print("número negativo")
```

```python
#atividade 02
numero = int(input("Digite seu número: "))
if numero % 2 == 0:
  print("o número e par")
else:
  print("o número é impar")
```

```python
#atividade 03
nota = float(input("Digite sua nota: "))
if nota >= 7:
  print("aprovado")
elif nota >= 5:
  print("Recuperação")
else:
  print ("Reprovado")
```
